```python
# keep the rows from df2 where the image_id is in common
annotations_common = annotations[annotations['image_id'].isin(common.index)]
annotations_common
```

```python
labels = ["text","title","list","table","figure"]

# add a new column with the category name
# annotations_common['category'] = annotations_common['category_id'].apply(lambda x: labels[x-1])
# annotations_common
```

```python
#group them by image id and keep the groups that do not have figures
annotations_common_no_figures = annotations_common.groupby(['image_id']).filter(lambda x: (x['category_id'] != 5).all())
print(len(annotations_common['image_id'].unique()))
print(len(annotations_common_no_figures['image_id'].unique()))
annotations_common_no_figures
```

```python
# add a new column with the image name
# df5 = annotations_common_no_figures.join(common, on='image_id')
# df5
```

```python
# divide the pages into train, val and test sets. First group then split 
from sklearn.model_selection import train_test_split

train, test = train_test_split(common, test_size=0.2, random_state=42)
train, val = train_test_split(train, test_size=0.15, random_state=42)

print(len(train))
print(len(val))
print(len(test))
```

```python
annotations_train = annotations_common_no_figures[annotations_common_no_figures['image_id'].isin(train.index)]
annotations_val = annotations_common_no_figures[annotations_common_no_figures['image_id'].isin(val.index)]
annotations_test = annotations_common_no_figures[annotations_common_no_figures['image_id'].isin(test.index)]
```

```python
# for each set, create a new dictionary with the images and the annotations and the categories (copied from the original dictionary)
train_dict = data.copy()
val_dict = data.copy()
test_dict = data.copy()

train_dict["images"] = train.to_dict(orient='records')
val_dict["images"] = val.to_dict(orient='records')
test_dict["images"] = test.to_dict(orient='records')

train_dict["annotations"] = annotations_train.to_dict(orient='records')
val_dict["annotations"] = annotations_val.to_dict(orient='records')
test_dict["annotations"] = annotations_test.to_dict(orient='records')
```

```python
import json
with open('train_syn.json', 'w') as f:
    json.dump(train_dict, f)
    
with open('val_syn.json', 'w') as f:
    json.dump(val_dict, f)
    
with open('test_syn.json', 'w') as f:
    json.dump(test_dict, f)
```
